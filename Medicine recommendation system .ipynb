```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

import warnings
warnings.filterwarnings('ignore')
```

```python
# Preprocessing
X = df.drop("prognosis", axis=1)
y = df["prognosis"]
```

```python
X.shape, y.shape
```

```python
# Label encode the target variable
le = LabelEncoder()
le.fit(y)
Y = le.transform(y)
```

```python
print(Y)
print(le.inverse_transform(Y))
```

```python
X_train, X_test, y_train , y_test = train_test_split(X,Y, test_size=0.3, random_state=42)
```

```python
X_train.shape, X_test.shape , y_train.shape, y_test.shape
```

```python
models = {
    "SVC" : SVC(kernel='linear'),
    "RandomForest" : RandomForestClassifier(n_estimators=100, random_state=42),
    "GradientBoost" : GradientBoostingClassifier(n_estimators=100, random_state=42),
    "MultinomialNB" : MultinomialNB(),
    "KNeighborsClassifier" : KNeighborsClassifier(n_neighbors=5)
}
```

```python
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)    #check accuracy 
    cun = confusion_matrix(y_test, y_pred)       #confusion matrix
    
    print('\n******************************************************************\n')
    print(f"{model_name} accuracy : {accuracy}")
    print(f"{model_name } Confusion_matrics :\n {cun}")
    print(np.array2string(cun, separator = ",") )
```

```python
rf = RandomForestClassifier(n_estimators= 100 , random_state=42)
rf.fit(X_train , y_train)
y_pred = rf.predict(X_test)
accuracy = accuracy_score(y_test , y_pred)
accuracy
```

```python
svc = SVC(kernel = 'linear')
svc.fit(X_train , y_train)
y_pred = svc.predict(X_test)
accuracy = accuracy_score(y_test , y_pred)
accuracy
```

```python
import pickle 
pickle.dump(rf , open('rf.pkl','wb'))     # for RF model 

pickle.dump(svc , open('svc.pkl','wb'))  # for Svc model 
```
